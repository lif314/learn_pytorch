```python
import torch
from torch import nn
from torch.nn import Conv2d, MaxPool2d, Flatten,Linear, Sequential
# Sequential  简化网络构建
```

```python
class CIFAR10_Model(nn.Module):
    def __init__(self):
        super().__init__()
        # 输入 3@32*32   输出 32@32*32
#         self.conv1 = Conv2d(3, 32, 5, padding=2,stride=1)
#         self.maxpool1 = MaxPool2d(2)
#         self.conv2 = Conv2d(32, 32, 5, padding=2)
#         self.maxpool2 = MaxPool2d(2)
#         self.conv3 = Conv2d(32, 64, 5, padding=2)
#         self.maxpool3 = MaxPool2d(2)
#         self.flattern = Flatten()  # 64*4*4=1024
#         self.linear1 = Linear(1024, 64)   # 线性层
#         self.linear2 = Linear(64, 10)  # 输出为10个类别
        
        # 使用Sequential创建网络
        self.sequential = Sequential(
             Conv2d(3, 32, 5, padding=2,stride=1),
             MaxPool2d(2),
             Conv2d(32, 32, 5, padding=2),
             MaxPool2d(2),
             Conv2d(32, 64, 5, padding=2),
             MaxPool2d(2),
             Flatten(),
             Linear(1024, 64),
             Linear(64, 10)
        )
        
    def forward(self, x):
        x = self.sequential(x)
#         x = self.conv1(x)
#         x = self.maxpool1(x)
#         x = self.conv2(x)
#         x = self.maxpool2(x)
#         x = self.conv3(x)
#         x = self.maxpool3(x)
#         x = self.flattern(x)
#         x = self.linear1(x)
#         x = self.linear2(x)
        return x
```

**计算卷积层的参数设置**



```python
# 网络实例
model = CIFAR10_Model()
print(model)
```

```python
# 检查网络的正确性
input = torch.ones((64, 3, 32,32))  # batch_size=64  输入
output = model(input)
print(output.shape)
```

```python
# 可视化网络结构
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("graphs")
writer.add_graph(model, input)
writer.close()
```
