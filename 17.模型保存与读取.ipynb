```python
import torchvision
from torch import  nn
```

```python
vgg16 = torchvision.models.vgg16(pretrained=False)
```

```python
print(vgg16)
```

```python
# 修改模型
vgg16.classifier[6] = nn.Linear(4096, 10)
```

```python
print(vgg16)
```

```python
# 保存模型--方法1
# 存在陷阱: 自定义网络保存后，加载时找不到类
import torch

# 保存网络模型结构和参数
torch.save( vgg16,r"./model/vgg16_model1.pth")
```

```python
# 加载模型
model = torch.load(r"./model/vgg16_model1.pth")
print(model)
```

```python
# 保存模型方法2
# 保存模型参数 【官方推荐】
torch.save(vgg16.state_dict(), "./model/vgg16_model2.pth")
```

```python
small_model = torch.load("./model/vgg16_model2.pth")
# 字典参数
print(small_model)
```

```python
# 新建网络模型结构 -- 加载网络模型
vgg16 = torchvision.models.vgg16(pretrained=False)
# 修改模型
vgg16.classifier[6] = nn.Linear(4096, 10)
# 加载模型参数
vgg16.load_state_dict(torch.load("./model/vgg16_model2.pth"))
print(vgg16)
```
